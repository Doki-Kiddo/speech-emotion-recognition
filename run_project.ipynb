```python
import os
import sys
from pathlib import Path

PROJECT_ROOT = Path.cwd()
print("Project root:", PROJECT_ROOT)
print("Python:", sys.executable)

if PROJECT_ROOT.name != "speech-emotion-cnn":
    raise RuntimeError("Open this notebook from the speech-emotion-cnn project folder.")

required = ["backend", "frontend", "src", "models", "data"]
missing = [name for name in required if not (PROJECT_ROOT / name).exists()]
if missing:
    raise RuntimeError(f"Missing project folders: {missing}")

if ".venv" not in sys.executable:
    print("WARNING: This notebook is not using the project .venv kernel.")
    print("Go to Kernel -> Change kernel -> Python (.venv speech-emotion-cnn), then restart.")
else:
    print("Using project .venv kernel.")
```

```python
from pathlib import Path
import json

from src.config import MODEL_PATH, LABELS_PATH

model_path = Path(MODEL_PATH)
labels_path = Path(LABELS_PATH)
print("Model exists:", model_path.exists(), model_path)
print("Labels exist:", labels_path.exists(), labels_path)

if not model_path.exists() or not labels_path.exists():
    raise RuntimeError("Model files are missing. Set TRAIN_MODEL = True in the next cell and run it.")

labels = json.loads(labels_path.read_text(encoding="utf-8"))
print("Labels:", labels)
```

```python
import subprocess
import sys

TRAIN_MODEL = False

if TRAIN_MODEL:
    result = subprocess.run([
        sys.executable,
        'train.py',
        '--data-dir', 'data/raw',
        '--epochs', '40',
        '--batch-size', '32',
    ], check=True)
else:
    print("Skipping training. Set TRAIN_MODEL = True to retrain.")
```

```python
import os
import subprocess
import sys

env = os.environ.copy()
env['TF_CPP_MIN_LOG_LEVEL'] = '2'

result = subprocess.run([
    sys.executable,
    'evaluate.py',
    '--data-dir', 'data/raw',
    '--split', 'validation',
    '--test-size', '0.2',
    '--seed', '42',
    '--report-only',
], check=True, capture_output=True, text=True, env=env)

print(result.stdout)
```
